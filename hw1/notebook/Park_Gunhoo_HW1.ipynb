````python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
train_error, test_error = {}, {}
````

````python
# default p is 2, which is euclidean_distance, when p=1, this is equivalent to suing manhattan_distance
# Soucre: https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
for k in range(208, 0, -3):
    knn = KNeighborsClassifier(n_neighbors=k, p=2)
    knn.fit(x_train, y_train)
    train_y_predict = knn.predict(x_train)
    test_y_predict = knn.predict(x_test)
    train_error[k] = np.mean(train_y_predict!=y_train)
    test_error[k] = np.mean(test_y_predict!=y_test)
````

#### ii. Calculate true positive rate, true negative rate, precision, and F1-score

````python
# The true positive rate(TPR) = TP/TP+FN
# The true negative rate(TNR) = TN/TN+FP
# precision = TP / (TP + FP)
# F1-score = 2*TP/(2*TP+FP+FN)
````

````python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
TPR = tp / (tp + fn)
TNR = tn / (tn + fp)
precision = tp / (tp + fp)
f1_score = 2*tp / (2*tp + fp + fn)
````
