# 1. Decision Trees as Interpretable Models

## (a) Download the Acute Inflamations data from https://archive.ics.uci.edu/dataset/184/acute+inflammations

```python
FILE_PATH ='../data/acute+inflammations/diagnosis.data'
HEADER = ['Temperature','Nausea','Lumbar','Urine','Micturition','Urethra','Inflammation','Nephritis']
df = pd.read_csv(FILE_PATH, encoding='UTF-16', sep='\t', names=HEADER, decimal=",")
df
```

## (b) Build a decision tree on the whole data set and plot it.

```python
df=df.replace({'no':0, 'yes':1})
FEATURES = ['Temperature','Nausea','Lumbar','Urine','Micturition','Urethra']
LABELS = ['Inflammation', 'Nephritis'] 

X = df[FEATURES]
y = df[LABELS]

# Source: https://scikit-learn.org/dev/modules/generated/sklearn.tree.DecisionTreeClassifier.html
clf = DecisionTreeClassifier(random_state=42)
clf = clf.fit(X, y)
```

```python
# Source: https://scikit-learn.org/1.5/modules/generated/sklearn.tree.plot_tree.html
plot_tree(clf, feature_names = FEATURES)
plt.show()
```

## (c) Convert the decision rules into a set of IF_THEN rules.

```python
# Source: https://www.kdnuggets.com/2017/05/simplifying-decision-tree-interpretation-decision-rules-python.html
# Source: https://stackoverflow.com/questions/20224526/how-to-extract-the-decision-rules-from-scikit-learn-decision-tree
def tree_to_code(tree, feature_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]
    print("def tree({}):".format(", ".join(feature_names)))

    def recurse(node, depth):
        indent = "  " * depth
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            print("{}if {} <= {}:".format(indent, name, threshold))
            recurse(tree_.children_left[node], depth + 1)
            print("{}else:  # if {} > {}".format(indent, name, threshold))
            recurse(tree_.children_right[node], depth + 1)
        else:
            print("{}return [{}\n\t{}{}]".format(indent, tree_.value[node][0]* tree_.weighted_n_node_samples[node], indent, tree_.value[node][1]* tree_.weighted_n_node_samples[node]))

    recurse(0, 1)
```

```python
tree_to_code(clf, FEATURES)
```

## (d) Use cost-complexity prunning to find a minimal decision tree and a set of decision rules with high interpretability.

```python
# Source: https://scikit-learn.org/1.5/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier.cost_complexity_pruning_path
path = clf.cost_complexity_pruning_path(X, y)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
plt.plot(ccp_alphas, impurities, marker='o', drawstyle="steps-post")
plt.xlabel("Effective alpha")
plt.ylabel("Total impurity of leaves")
plt.title("Cost-Complexity Pruning Path")
plt.grid(True)
plt.show()
```

```python
grid_search_cv = GridSearchCV(tree.DecisionTreeClassifier(), param_grid={'ccp_alpha': ccp_alphas}, cv=5, scoring='accuracy')
grid_search_cv.fit(X, y)

grid_search_cv_alpah, grid_search_cv_score, grid_search_cv_std = grid_search_cv.cv_results_['param_ccp_alpha'], grid_search_cv.cv_results_['mean_test_score'], grid_search_cv.cv_results_['std_test_score']

df_results = pd.DataFrame({
    'ccp_alpha': grid_search_cv.cv_results_['param_ccp_alpha'],
    'mean_test_score': grid_search_cv.cv_results_['mean_test_score'],
    'std_test_score': grid_search_cv.cv_results_['std_test_score']
})

df_results
```

```python
filtered_df = df_results[df_results['ccp_alpha'] > 0.0]
best_row = filtered_df.loc[filtered_df['mean_test_score'].idxmax()]

print("The highest test score is {}, so we pick ccp_alpha = {}".format(best_row['mean_test_score'], best_row['ccp_alpha']))
```

The tree overfits when ccp_alpha is set to zero while maintaining the other DecisionTreeClassifier default parameters. That is, when the ccp_alpha is zero, the tree is unpruned.  
Ref: https://scikit-learn.org/1.5/auto_examples/tree/plot_cost_complexity_pruning.html

```python
best_ccp_alpha = best_row['ccp_alpha']
clf_best = DecisionTreeClassifier(random_state=42, ccp_alpha=best_ccp_alpha)
clf_best.fit(X, y)
```

```python
plot_tree(clf_best, feature_names=FEATURES)
plt.show()
```

```python
tree_to_code(clf_best, FEATURES)
```

# 2. The LASSO and Boosting for Regression

## (a) Download the Communities and Crime data from https://archive.ics.uci.edu/ml/datasets/Communities+and+Crime. Use the first 1495 rows of data as the training set and the rest as the test set.

```python
NAMES_FILEPATH = '../data/acute+inflammations/communities.names'
COMMUNITIES_FILEPATH = '../data/acute+inflammations/communities.data'

with open(NAMES_FILEPATH, "r") as f:
    column_names = []
    for line in f:
        words = line.strip().split()
        if words and words[0] == "@attribute":
            column_names.append(words[1]) 

df = pd.read_csv(COMMUNITIES_FILEPATH, names=column_names, na_values="?")
df
```

```python
training_set = df[:1495]
testing_set = df[1495:]
```

## (b) The data set has misssing values. Use a data imputation technique to deal with the missing values in the data set. The data-description mentions some features are nonpredictive. Ignore those features.

```python
nonpredictive=['state','county','community','communityname','fold']
training_set.drop(nonpredictive, axis=1, inplace=True)
testing_set.drop(nonpredictive, axis=1, inplace=True)
training_set.replace('?', np.nan)
testing_set.replace('?', np.nan)
```

```python
# Source: https://scikit-learn.org/1.5/modules/generated/sklearn.impute.SimpleImputer.html
imputer = SimpleImputer(missing_values = np.nan, strategy='mean')
imputer.fit(training_set)
training_set = pd.DataFrame(imputer.transform(training_set), columns=training_set.columns)

imputer = SimpleImputer(missing_values = np.nan, strategy='mean')
imputer.fit(testing_set)
testing_set = pd.DataFrame(imputer.transform(testing_set), columns=testing_set.columns)
```

## (c) Plot a correlation matrix for the features in the data set.

```python
df = pd.concat([training_set, testing_set], ignore_index=True)
correlation_matrix = df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, fmt='.2f', cbar=True)
plt.title("Correlation Matrix")
plt.show()
```

## (d) Calculate the Coefficient of Variation CV for each feature, where CV = s/m, in which s is sample standard deviation and m is sample mean

```python
mean_values = df.mean()
std_values = df.std()
cv_values = std_values / mean_values

cv_values = cv_values.to_frame(name="CV")
cv_values
```

## (e) Pick |âˆš128| features with highest CV , and make scatter plots and box plots for them. Can you draw conclusions about significance of those features, just by the scatter plots?ures, just by the scatter plots?

```python
top_cv = cv_values.sort_values(ascending = False, by=['CV'])[:math.floor(math.sqrt(128))]
top_cv
```

```python
sns.pairplot(df[list(top_cv.index)])
plt.suptitle("Scatter Plot Matrix of Top Features by CV", y=1.02)
plt.show()
```

## (f) Fit a linear model using least squares to the training set and report the test error.

```python
X_train = training_set.iloc[:, :-1]
y_train = training_set.iloc[:, -1]
X_test = testing_set.iloc[:, :-1]
y_test = testing_set.iloc[:, -1]
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

test_error = mean_squared_error(y_test, y_pred)
test_error
```

## (g) Fit a ridge regression model on the training set, with Î» chosen by cross-validation. Report the test error obtained.

```python
# Source: https://scikit-learn.org/dev/modules/generated/sklearn.linear_model.RidgeCV.html
alpha_values = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

ridge_cv = RidgeCV(alphas=alpha_values, store_cv_values=True)
ridge_cv.fit(X_train, y_train)
```

```python
chosen_lambda = ridge_cv.alpha_
y_pred_ridge = ridge_cv.predict(X_test)

test_error_ridge = mean_squared_error(y_test, y_pred_ridge)

print(f"With chosen lamda: {chosen_lambda}")
print('test error: ', test_error_ridge)
```

## (h) Fit a LASSO model on the training set, with Î» chosen by cross-validation. Report test error obtained, along with a list of the variables selected by the model. Repeat with standardized features. Report the test error for both cases and compare them

```python
# Lasso without standardization
```

```python
lasso_cv = LassoCV(cv=5, random_state=0) 
lasso_cv.fit(X_train, y_train)
```

```python
y_pred_lasso = lasso_cv.predict(X_test)
test_error_lasso = mean_squared_error(y_test, y_pred_lasso)
selected_features = X_train.columns[(lasso_cv.coef_ != 0)]

print(f"Selected features using Lasso without standardization: {selected_features}")
print('test error: ', test_error_lasso)
```

```python
y_pred_lasso_std = lasso_cv_std.predict(X_test_std)
test_error_lasso_std = mean_squared_error(y_test, y_pred_lasso_std)
selected_features_std = X_train.columns[(lasso_cv_std.coef_ != 0)]

print(f"Selected features using Lasso using standardization: {selected_features_std}")
print('test error: ', test_error_lasso_std)
```

```python
print(f'Test error using lasso without standardization is {test_error_lasso}, while test error using lasso with standardizatino is {test_error_lasso_std}. This shows there is only few differences bewteen them {test_error_lasso_std - test_error_lasso}.')
```

## (i) Fit a PCR model on the training set, with M (the number of principal components) chosen by cross-validation. Report the test error obtained.

```python
m_values = range(1, X_train.shape[1] + 1)
cv_errors = []
```

```python
# Source: https://scikit-learn.org/dev/modules/generated/sklearn.decomposition.PCA.html
for m in m_values:
    pca = PCA(n_components=m)
    X_train_pca = pca.fit_transform(X_train)
    mse_cv = -cross_val_score(model, X_train_pca, y_train, cv=5, scoring='neg_mean_squared_error').mean()
    cv_errors.append(mse_cv)

best_m = m_values[np.argmin(cv_errors)]
print(f"Optimal M is {best_m}")
```

```python
pca = PCA(n_components=best_m)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

final_model = LinearRegression()
final_model.fit(X_train_pca, y_train)

y_pred_pcr = final_model.predict(X_test_pca)
test_error_pcr = mean_squared_error(y_test, y_pred_pcr)

print("Test Error (MSE) with PCR:", test_error_pcr)
```

```python
plt.plot(m_values, cv_errors, linestyle='-', color='b', label='Cross-validation error')
plt.axvline(x=best_m, color='r', linestyle='--', label=f'Optimal M = {best_m}')
plt.xlabel('Number of Principal Components (M)')
plt.ylabel('Cross-Validation Mean Squared Error')
plt.title('Cross-Validation Error vs. Number of Principal Components')
plt.legend()
plt.grid(True)
plt.show()
```

## (j) In this sections, we would like to fit a boosting tree to the data. As in classification trees, one can use any type of regression at each node to build a multivariate regression tree. Because the number of variables is large in this problem, one can use L1-penalized regression at each node. Such a tree is called L1 penalized gradient boosting tree. You can use XGBoost to fit the model tree. Determine alpha(the regularization term) using cross-validation.

```python
# Source: https://xgboost.readthedocs.io/en/latest/python/python_api.html#xgboost.XGBRegressor
alpha_params = {'alpha':alpha_values}

xgb_regressor = XGBRegressor(objective='reg:squarederror', n_estimators=100, random_state=42)

grid_search = GridSearchCV(estimator=xgb_regressor, param_grid=alpha_params, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

best_alpha = grid_search.best_params_['alpha']
print(f'Best alpha is {best_alpha}')

xgb_final = XGBRegressor(objective='reg:squarederror', n_estimators=100, alpha=best_alpha, random_state=42)
xgb_final.fit(X_train, y_train)

y_pred = xgb_final.predict(X_test)
test_error = mean_squared_error(y_test, y_pred)

print(f'Test error with the best alpha ({best_alpha}) is {test_error}')
```
